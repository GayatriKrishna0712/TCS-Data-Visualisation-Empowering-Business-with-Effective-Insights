Q1. The CEO of the retail store is interested to view the time series of the revenue data for the entire year. The CEO is interested in viewing the seasonal trends and wants to dig deeper into why these trends occur. This analysis will be helpful for the CEO to forecast for the next year.

```python
df['Year'] = pd.DatetimeIndex(df['InvoiceDate']).year
df.head()
```

```python
df.drop(df[df['Country'] == 'Unspecified'].index, inplace = True)
```

```python
from datetime import datetime
import calendar
```

```python
df['Month'] = df['InvoiceDate'].dt.month_name(locale='English')
df.head()
```

```python
print("The variable type: \n",df.info())
```

```python
df["TotalPrice"] = df['Quantity']*df['UnitPrice']
df.head()
```

```python
#unique values from each col
for i in df.columns:
    print(i, "column, unique values: ", df[i].nunique())
    print("************")
    print("")
```

```python
#checking if there exist quantity or price less than 0 
df.loc[(df["Quantity"]<=0) | (df["UnitPrice"]<=0)]
```

```python
a = df[['Quantity','UnitPrice','Description','Year','TotalPrice','Month']]
a.head()
```

```python
a.nunique()
```

```python
g1 = pd.DataFrame(a.groupby(['Year','Month'])['TotalPrice'].mean().reset_index())
print(g1)
g1.plot(kind = 'line',color='red',x = 'Month',y = 'TotalPrice')
```

```python
g2 = pd.DataFrame(a.groupby(['Year']).TotalPrice.mean())
print(g2)
g2.plot(kind='bar')
```

Q2. The CMO is interested in viewing the top 10 countries which are generating the highest revenue. Additionally, as a subcomponent, they would also like to see which products are contributing to the total revenue being generated by each country.

```python
g3 = pd.DataFrame(df.groupby(['Country'])['TotalPrice'].sum().reset_index())
g3 = g3.sort_values('TotalPrice', ascending=False)
g3 = g3.head(10)
print(g3)
g3.plot(kind='barh',x = 'Country',y = 'TotalPrice',stacked=True)
```

```python
g3['Description'] = df['Description']
g3
```

```python
pd.DataFrame(df.groupby(['Country','Description'])['TotalPrice'].mean().reset_index())
```

Q3. The CEO of the online retail store wants to see how much average revenue is generated by each country. They are interested in viewing the following metrics on the visual:
* Minimum value
* First quartile value
* Median value 
* Third quartile value
* Maximum value

```python
g5 = pd.DataFrame(df.groupby(['Country'])['TotalPrice'].mean().reset_index())
g5
```

```python
plt.figure(figsize=(35,35))
sns.boxplot(x="TotalPrice", y="Country", data=g5)
plt.ylim(-0.5, 9.5)
plt.show()
```
